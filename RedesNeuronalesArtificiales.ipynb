## Arquitectura de Redes  Neuronales

Se denomina arquitectura a la estructura (topología) en la que las distintas neuronas constituyentes de la red neuronal se asocian.

Las neuronas se suelen agrupar en unidades estructurales denominadas capas. Dentro de una capa, las neuronas pueden agruparse formando grupos neuronales.

El conjunto de una o más capas constituye una red neuronal.



Podemos distinguir tres tipos de capas: (Ver la figura siguente)

*  Capa de entrada:   
    Compuesta por neuronas que reciben datos o señales procedentes del entorno.
*  Capa oculta:  
    Aquellas que no tiene una conexión directa con el entorno.  
*  Capa de salida:  
    Aquellaa cuyas neuronas proporcionan la respuesta de la red neuronal.


image.png

Las conexiones pueden clasificarse también en conexiones _intracapa_  y conexiones _intercapa_.

Las primeras se corresponden con las conexiones entre las neuronas de una misma capa y, la segunda se corresponde a las conexiones entre neuronas de distintas capas.

##  Tipos de Redes Neuronales

Enumeramos las diferentes estructuras en las que las neuronas se pueden asociar.



### Red Neuronal Monocapa (Perceptrón)


Se corresponde con la red neuronal más sencilla, ya que se tiene una capa de neuronas que proyectan las entradas a una capa de neuronas de salida donde se realizan los diferentes cálculos. (Obervar la siguiente Figura inciso a)

Algunos ejemplos de aplicaciones para redes neuronales monocapas:

__Operaciones Matemáticas Simples__: Puedes utilizar una red monocapa para aprender operaciones matemáticas simples como suma o multiplicación.

__Detección de Fraudes__: En aplicaciones financieras, una red monocapa puede utilizarse para detectar transacciones fraudulentas. Toma características de una transacción, como el monto y la ubicación, y decide si es fraudulenta o no.

__Clasificación de Texto__: En tareas de clasificación de texto binario, una red monocapa puede determinar si un texto pertenece a una categoría o no. Por ejemplo, clasificar si un correo electrónico es spam o no spam.

__Filtrado de Ruido__: En procesamiento de señales, una red monocapa podría usarse para filtrar ruido en una señal.

__Control de Robots Simples__: En robótica, una red monocapa podría utilizarse para controlar robots simples que toman decisiones basadas en entradas sensoriales.

__Clasificación de Imágenes Simples__: Aunque no es común, una red monocapa puede utilizarse para tareas muy simples de clasificación de imágenes, como distinguir entre imágenes de gatos y perros en blanco y negro.

### Red Neuronal Multicapa

Es una generalización de la anterior, existiendo un conjunto de capas intermedias entre la capa de entrada y la de salida (capas ocultas).


Este tipo de red puede estar total o parcialmente conectada. (Obervar la siguiente Figura inciso b)

Algunos ejemplos de aplicaciones de redes neuronales multicapas:

__Reconocimiento de Patrones en Imágenes__: Las redes neuronales convolucionales (CNN) son un tipo de red multicapa ampliamente utilizadas en tareas de visión por computadora, como la clasificación de objetos en imágenes, la detección de objetos, el etiquetado de imágenes y la segmentación semántica.

__Recomendación de Contenido__: Las redes neuronales se utilizan para crear sistemas de recomendación que sugieren productos, películas, música y contenido personalizado a los usuarios.

__Detección de Anomalías__: En aplicaciones de seguridad cibernética y monitoreo de sistemas, las redes neuronales multicapas se utilizan para detectar anomalías y comportamientos sospechosos en los datos.

__Diagnóstico Médico__: Las redes neuronales se aplican en el procesamiento de imágenes médicas para la detección temprana de enfermedades, como el cáncer, y en la clasificación de imágenes de resonancia magnética y tomografía computarizada.

__Control de Juegos__: En la inteligencia artificial, las redes neuronales se utilizan para entrenar agentes de IA en juegos como ajedrez, Go y videojuegos.

__Traducción Automática__: Las redes neuronales, como las redes neuronales basadas en transformadores, se utilizan en sistemas de traducción automática para traducir texto entre diferentes idiomas.

### Red Neuronal Recurrente (RNN)

Este tipo de red se diferencia de las anteriores en la existencia de lazos de realimentación en la red. Estos lazos pueden ser entre neuronas de diferentes capas, neuronas de la misma capa o, entre una misma neurona. (Obervar la siguiente Figura inciso c)


Esta estructura la hace especialmente adecuada para estudiar la dinámica de los sistemas no lineales.

Algunos ejemplos de aplicaciones de redes neuronales recurrentes:

__Procesamiento de Lenguaje Natural (NLP)__:  
  
  _Traducción Automática_: Las RNN, especialmente las LSTM y GRU, se utilizan en sistemas de traducción automática para convertir texto de un idioma a otro.  
  
  _Generación de Texto_: Las RNN se utilizan para generar texto coherente y creativo, como en la generación de historias, poesía y diálogos.

  _Modelado de Lenguaje_: Las RNN se aplican para modelar la probabilidad de una secuencia de palabras, lo que es útil en tareas de autocompletado y corrección gramatical.

__Predicción de Series Temporales__: Las redes neuronales, en particular las redes LSTM(Long Short Time Memory), se utilizan en la predicción de series temporales en aplicaciones financieras, meteorológicas y de inventario.

__Reconocimiento de Voz__: Las RNN se utilizan en sistemas de reconocimiento de voz para convertir señales de audio en texto.

__Análisis de Sentimientos__: En el procesamiento de lenguaje natural, las RNN se utilizan para clasificar el sentimiento de un texto, como determinar si una reseña de película es positiva o negativa.

__Predicción de Series Temporales__: Las RNN, en particular las LSTM y GRU, se utilizan para predecir series temporales, como pronósticos meteorológicos, precios de acciones y demanda de productos.

__Generación de Música__: Las RNN se utilizan en la generación de música, donde la red aprende patrones temporales en la música y crea composiciones musicales.
Modelado de Trayectorias:

__Chatbots e Interacción de Diálogo__: Las RNN se utilizan en sistemas de chatbots para mantener un contexto en las conversaciones y generar respuestas más coherentes.
Etiquetado de Secuencias:

__Predicción de Texto con Autocompletado__: Las RNN se utilizan en sistemas de autocompletado de texto, como en las sugerencias de búsqueda de motores de búsqueda y aplicaciones de mensajería.
